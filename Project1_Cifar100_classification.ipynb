```python
import tensorflow as tf
from tensorflow.keras import layers, models
import tensorflow_datasets as tfds
```

## Importing Dataset

```python
dataset_name = 'cifar100'
(train_dataset, test_dataset), info = tfds.load(name=dataset_name, split=['train', 'test'], with_info=True)
```

## Data Preprocessing

```python
# Preprocess the data
def preprocess(example):
    image = tf.image.resize(example['image'], (32, 32))  # Resize images to a consistent size
    image = tf.cast(image, tf.float32) / 255.0  # Normalize pixel values to [0, 1]
    label = tf.cast(example['label'], tf.int32)  # Convert labels to int32
    return image, label
```

```python
train_dataset = train_dataset.map(preprocess)
test_dataset = test_dataset.map(preprocess)
```

## Model Building

```python
# Build the CNN model
model = models.Sequential()

model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(100, activation='softmax'))  # 100 classes for CIFAR-100
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

## Model Training

```python
# Train the model
history = model.fit(train_dataset.batch(64), epochs=10, validation_data=test_dataset.batch(64))
```

## Model Evaluation

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_dataset.batch(64))
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```
